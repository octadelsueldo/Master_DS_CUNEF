```python
## Importaciones

import nltk, re, pprint
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer

# Importamos las librerías que corresponda
# Hay que añadir los modelos que se utilicen. Previamente cargados con python -m spacy download es_core_news_md

import spacy
from spacy import displacy
import pandas as pd
```

```python
## Leemos el archivo en ingles

engl=open("/Users/octaviodelsueldo/Documents/CUNEF/DS para la Gestion de Informacion no Estructurada/Actividades/Actividad 1. POS y Extracciçon de Entidades/ing.txt","r").read()
print(engl)
```

```python
## Separación de frases

print("Separación de frases en Inglés")
sent_tokenizer=nltk.data.load('tokenizers/punkt/english.pickle')

sents = sent_tokenizer.tokenize(engl)
for sent in sents:
    print(sent)
```

El separador de NLTK lo hace muy bien en Inglés separando las frases hasta el punto seguido perfectamente.

Vamos a analizar ahora, con el paquete NLTK, la tokenizacion o dividision de un texto en las unidades más pequeñas que tengan significado, es decir en palabras.

```python
## Tokenización del texto en Inglés

print("Tokens English")
tokens_engl = nltk.word_tokenize(engl)
print(tokens_engl[:15])
```

Podemos observar para los primeros 15 tokens reconocidos que en inglés tiene una perfecta identificación de las palabras


Veamos que sucede ahora al transformar los tokens a su forma canonica

```python
## Stemming

stemmer=SnowballStemmer('english')
for token in tokens_engl:
    print(stemmer.stem(token))
```

En ingles, funciona muy bien salvo por algunas palabras como "history" que el algoritmo se confunde y transforma la palabra a histori. Tambien la palabra discriminate esta mal transformada.

Veamos ahora el análisis morfológico donde iremos tokenizando y categorizando las palabras que consituyen el texto según su categoría morfológica (nombre, verbo...)

```python
## Part of Speech (Análisis morfológico)

# Tokenizamos por frases, no todo el texto

tokens = [nltk.word_tokenize(sent) for sent in sents]

tokens_tagged = [nltk.pos_tag(token) for token in tokens]
print(tokens_tagged)
```

```python
## Extracción de entidades

# Haremos un análisis adicional sobre los tokens de los que hemos hecho un análisis morfológico

for token in tokens_tagged:
    print(nltk.ne_chunk(token))
```

```python
# Leemos el archivo en español y comparamos

espa=open("/Users/octaviodelsueldo/Documents/CUNEF/DS para la Gestion de Informacion no Estructurada/Actividades/Actividad 1. POS y Extracciçon de Entidades/esp.txt", "r").read()
print(espa)
```

### Vamos a comparar ahora con el paquete spacy

```python
# Cargamos los modelos en ingles y en español

nlp_es = spacy.load('es_core_news_md')
nlp_eng = spacy.load('en_core_web_md')
```

```python
# Ejecutamos la pipeline de análisis

docum_analizado = nlp_es(espa)
docum_analyzed = nlp_eng(engl)
```

```python
# Imprimimos frases

for sent in docum_analizado.sents:
    print(sent.text)
for sent_eng in docum_analyzed.sents:
    print(sent_eng.text)
```

A la hora de separar en frases spicy funciona igual de bien, al menos en español e inglés, que NLTK.

Vamos ahora separando en tokens, formas canonicas y entidades

```python
# Imprimimos tokens, formas canónicas y entidades reconocidas

# POS tagging with Spacy 
spacy_pos_tagged = [(palabra.text, palabra.lemma_, palabra.pos_) for palabra in docum_analizado]
pd.DataFrame(spacy_pos_tagged, columns=['Palabra', 'F. Canónica', 'POS']).T
```

De un primer vistazo ya se puede apreciar la diferencia, sobre todo en el español, a la hora de analizar los textos. Spicy mejora la precisión de NLTK al acertar en la forma canónica de las palabras o tokens al mismo tiempo que presenta los resultados de una manera más clara y visual.

```python
# Tokens, formas canónicas y POS en inglés

spacy_eng_pos_tg = [(word.text, word.lemma_, word.pos_) for word in docum_analyzed]
pd.DataFrame(spacy_eng_pos_tg, columns=['Palabra', 'F_Canónica', 'POS']).T
```

En inglés, los resultados de las transformaciones son perfectos en este ejemplo.

```python
# Imprimimos entidades

ents_tagged = [(ent.text, ent.label_) for ent in docum_analizado.ents]
pd.DataFrame(ents_tagged, columns=['Entidad', 'Tipo']).T
```

```python
# Imprimimos entidades en inglés

ents_tagged = [(ent.text, ent.label_) for ent in docum_analyzed.ents]
pd.DataFrame(ents_tagged, columns=['Entidad', 'Tipo']).T
```

Veamos ahora el análisis de las entidades en el texto

```python
# Entidades en el texto

for sent in docum_analizado.sents:
    displacy.render(sent, style='ent', jupyter=True, options={'distance': 110,'arrow_stroke': 2,'arrow_width': 8})

```

```python
# Entidades en el texto en inglés

for sent in docum_analyzed.sents:
    displacy.render(sent, style='ent', jupyter=True, options={'distance': 110,'arrow_stroke': 2,'arrow_width': 8})

```

Cómo conclusión podemos decir que spicy es más útil para el análisis morfológico de textos sobre todo en español pero tambien en inglés no solo porque muestra mejor precisión en las transformaciones sino porque muestra los resultados en una manera más clara que facilita la interpretación.
